```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(['Female', 'Difference in Earnings between Physicians who Feels Fairly vs Unfairly Paid'], axis=1, inplace = True)
```

```python
le = LabelEncoder()
df['Specialty'] = le.fit_transform(df['Specialty'])
df['Feel Fairly Compensated'] = le.fit_transform(df['Feel Fairly Compensated'])
df['Overall Career Satisfaction'] = le.fit_transform(df['Overall Career Satisfaction'])
df['Satisfied w/ Income'] = le.fit_transform(df['Satisfied w/ Income'])
df['Would Choose Medicine Again'] = le.fit_transform(df['Would Choose Medicine Again'])
df['Would Choose the Same Specialty'] = le.fit_transform(df['Would Choose the Same Specialty'])
df['Survey Respondents by Specialty'] = le.fit_transform(df['Survey Respondents by Specialty'])
```

```python
sns.pairplot(df)
```

```python
x = df.drop(['Annual Income'], axis = 1)
y = df['Annual Income']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
####### Linear Regression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
y_train_pred = reg.predict(x_train)
y_test_pred = reg.predict(x_test)
```

```python
y_train_pred[:5]
```

```python
y_test_pred[:5]
```

```python
print(r2_score(y_train, y_train_pred)*100)
print(mean_squared_error(y_train, y_train_pred))
print(r2_score(y_test, y_test_pred)*100)
print(mean_squared_error(y_test, y_test_pred))
```

```python
###### Random Forest
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rf.fit(x_train, y_train)
```

```python
y_train_pred = rf.predict(x_train)
y_test_pred = rf.predict(x_test)
```

```python
print(r2_score(y_train, y_train_pred)*100)
print(mean_squared_error(y_train, y_train_pred))
print(r2_score(y_test, y_test_pred)*100)
print(mean_squared_error(y_test, y_test_pred))
```

```python
#### Decision Tree
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(random_state=42)
dtr.fit(x_train, y_train)
```

```python
y_train_pred = dtr.predict(x_train)
y_test_pred = dtr.predict(x_test)
```

```python
print(r2_score(y_train, y_train_pred)*100)
print(mean_squared_error(y_train, y_train_pred))
print(r2_score(y_test, y_test_pred)*100)
print(mean_squared_error(y_test, y_test_pred))
```

```python
#### xgboost
```

```python
import xgboost as xb
```

```python
xg_reg = xb.XGBRegressor()
xg_reg.fit(x_train, y_train)
```

```python
y_train_pred = xg_reg.predict(x_train)
y_test_pred = xg_reg.predict(x_test)
```

```python
print(r2_score(y_train, y_train_pred)*100)
print(mean_squared_error(y_train, y_train_pred))
print(r2_score(y_test, y_test_pred)*100)
print(mean_squared_error(y_test, y_test_pred))
```
